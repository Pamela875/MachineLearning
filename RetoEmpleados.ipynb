##Correlaciones y PCA

```python
#Como atrition solo fueron 2 variables borramos una, ya que la otra indica ambos valores y la renombramos
EmpleadosAttrition = EmpleadosAttrition.drop(['Attrition_Yes'], axis=1)
EmpleadosAttrition = EmpleadosAttrition.rename(columns={'Attrition_No':'Attrition'})
```

```python
#Correlaciones
# Lista de variables
variables = [
    'BusinessTravel_Non-Travel', 'BusinessTravel_Travel_Frequently',
    'BusinessTravel_Travel_Rarely', 'BusinessTravel_nan',
    'Department_Human Resources', 'Department_Research & Development',
    'Department_Sales', 'EducationField_Human Resources',
    'EducationField_Life Sciences', 'EducationField_Marketing',
    'EducationField_Medical', 'EducationField_Other',
    'EducationField_Technical Degree', 'Gender_Female', 'Gender_Male',
    'JobRole_Healthcare Representative', 'JobRole_Human Resources',
    'JobRole_Laboratory Technician', 'JobRole_Manager',
    'JobRole_Manufacturing Director', 'JobRole_Research Director',
    'JobRole_Research Scientist', 'JobRole_Sales Executive',
    'JobRole_Sales Representative', 'MaritalStatus_Divorced',
    'MaritalStatus_Married', 'MaritalStatus_Single', 'MaritalStatus_nan',
    'Attrition'
]

# Calcular correlación respecto a Attrition_No
correlaciones = EmpleadosAttrition[variables].corr()["Attrition"]
```

```python
correlaciones
```

```python
#Correlaciones mayor o igual a 0.1
EmpleadosAttritionFinal = EmpleadosAttrition[['Attrition','BusinessTravel_Non-Travel','EducationField_Technical Degree','JobRole_Healthcare Representative',
                                              'JobRole_Laboratory Technician','JobRole_Research Director','JobRole_Sales Representative',
                                              'MaritalStatus_Divorced','MaritalStatus_Single']]
```

```python
EmpleadosAttritionFinal.columns
```

```python
#PCA
from sklearn.decomposition import PCA

pca = PCA() #instancia PCA se le pone un numero de acuerdo con las variables principales que se requieran, si no es nada son todas
pca.fit(EmpleadosAttritionFinal) #datos componentes principales
print(pca.components_)
```

```python
#Evaluación de componentes obtenidos
print(pca.explained_variance_)
```

Eso significa que, de la suma de esa lista, el primer componente explica 27.34 puntos de la varianza y el segundo 16.51. Es decir, que información de este conjunto de datos puede ser obtenida con cada componente, entre menor puntos no contiene mucha información.

```python
#Para pasar a porcentajes con..
print(pca.explained_variance_ratio_)
```

```python
#Formar nuevas variables
EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
```

```python
EmpleadosAttritionPCA
```

```python
#  Varianza acumulada
varianza_acum = np.cumsum(pca.explained_variance_ratio_) # % de la varianza de cada componente se va acumulando

# Número mínimo de componentes que explican al menos el 80%
n_componentes = (varianza_acum >= 0.80).argmax() + 1 #1 por los componentes que se cuentan desde 1
print("Número de componentes necesarios:", n_componentes)

# Agregar columnas C0, C1, ... al DataFrame final
for i in range(n_componentes):
    EmpleadosAttritionFinal = EmpleadosAttritionFinal.assign(**{f"C{i}": EmpleadosAttritionPCA[:, i]})

print(EmpleadosAttritionFinal.head())
```
